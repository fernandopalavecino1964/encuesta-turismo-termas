## Generación de datos sintéticos

Genera datos sintéticos para cada columna basándose en el análisis del paso anterior y en las distribuciones y tipos de datos observados.


```python
import numpy as np
from datetime import datetime, timedelta

# Generate synthetic data for each column
n_samples = 3500

# fecha (Date)
# Analyze the date range from the original data
df['fecha'] = pd.to_datetime(df['fecha'], format='%d/%m/%Y %H:%M:%S')
min_date = df['fecha'].min()
max_date = df['fecha'].max()
date_range = (max_date - min_date).days

synthetic_fecha = [min_date + timedelta(days=np.random.randint(0, date_range),
                                       seconds=np.random.randint(0, 86400))
                   for _ in range(n_samples)]

# Categorical columns (using value counts for distribution)
categorical_cols = ['origen', 'edad', 'motivo', 'transporte', 'dias', 'acompañante', 'frecuencia', 'alojamiento']
synthetic_data = {}

for col in categorical_cols:
    value_counts = df[col].value_counts(normalize=True)
    synthetic_data[col] = np.random.choice(value_counts.index, size=n_samples, p=value_counts.values)

# Store synthetic data in variables
synthetic_origen = synthetic_data['origen']
synthetic_edad = synthetic_data['edad']
synthetic_motivo = synthetic_data['motivo']
synthetic_transporte = synthetic_data['transporte']
synthetic_dias = synthetic_data['dias']
synthetic_acompañante = synthetic_data['acompañante']
synthetic_frecuencia = synthetic_data['frecuencia']
synthetic_alojamiento = synthetic_data['alojamiento']

# Display the first few synthetic dates to verify
display(synthetic_fecha[:5])

# Display value counts for a couple of synthetic categorical columns to verify distribution
display(pd.Series(synthetic_origen).value_counts(normalize=True))
display(pd.Series(synthetic_edad).value_counts(normalize=True))
```

## Creación del dataframe sintético

### Subtask:
Combina los datos generados en un nuevo DataFrame.


**Reasoning**:
Combine the generated synthetic data into a dictionary and create a new DataFrame from it.



```python
# Create a dictionary from the synthetic data
synthetic_data_dict = {
    'fecha': synthetic_fecha,
    'origen': synthetic_origen,
    'edad': synthetic_edad,
    'motivo': synthetic_motivo,
    'transporte': synthetic_transporte,
    'dias': synthetic_dias,
    'acompañante': synthetic_acompañante,
    'frecuencia': synthetic_frecuencia,
    'alojamiento': synthetic_alojamiento
}

# Create the synthetic DataFrame
df_synthetic = pd.DataFrame(synthetic_data_dict)

# Display the head of the synthetic DataFrame
display(df_synthetic.head())
```
