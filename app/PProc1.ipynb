```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 7))
sns.countplot(data=df_1, x='frecuencia', hue='edad', palette='viridis')
plt.title('Distribución de Frecuencia de Visita por Edad')
plt.xlabel('Frecuencia')
plt.ylabel('Conteo')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Edad')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Identify categorical columns (excluding 'fecha' which is a timestamp and 'acompañante' with missing values)
categorical_cols = df_1.select_dtypes(include='object').columns.tolist()
# Removing 'fecha' as it's a timestamp and not suitable for a simple count plot
if 'fecha' in categorical_cols:
    categorical_cols.remove('fecha')

# Generate count plots for each categorical column
for col in categorical_cols:
    plt.figure(figsize=(6, 3))
    ax = sns.countplot(data=df_1, y=col, order=df_1[col].value_counts().index, palette='viridis', hue=col, legend=False)
    plt.title(f'Distribución de {col}')
    plt.xlabel('Conteo')
    plt.ylabel(col)

    # Add percentage labels
    total = len(df_1[col])
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_width()
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y), ha='left', va='center')

    plt.tight_layout() # Adjust layout to prevent labels overlapping
    plt.savefig(f'{col}_distribution.png') # Save the figure
    plt.close() # Close the figure to free memory

print("Las visualizaciones han sido exportadas como archivos PNG.")
```
